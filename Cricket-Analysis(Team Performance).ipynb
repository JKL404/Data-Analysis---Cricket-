# Analyzing Winning and Lossing 



```python
#DataFrame containg 4 columns : Total Matches, Win, Lost, Tied/NR
teams_data1 = teams_data[['TotalMatches','Win','Lost','Tied/NR']]
teams_data1[['Win','Lost']].plot.bar(width=0.7, rot=30)

plt.xlabel('Teams')
plt.ylabel('Win-Loss')
plt.title('Comparision of Total Wins Vs Total Loss')
plt.style.use('ggplot')

for i,v in enumerate(teams_data1.Win):
    plt.text(i,v,v, fontsize='8')
for i,v in enumerate(teams_data1.Lost):
    plt.text(i,v,v, fontsize='8')
plt.show()
```

```python
#Teams with most number of Tied Matches and NR Matches
plt.subplot(3,1,1)
teams_data['Tied'].plot.bar(rot=30, width=0.4)
plt.xlabel('Teams',fontsize=8)
plt.ylabel('Tied Matches',fontsize=8)
plt.xticks(fontsize=8)
plt.title('Total Number of Tied Matches',fontsize=10)
for i,v in enumerate(teams_data['Tied']):
    plt.text(i,v,v,fontsize=8)

plt.subplot(3,1,3)
teams_data['NR'].plot.bar(rot=30, width=0.4)
plt.xlabel('Teams',fontsize=8)
plt.ylabel('No Result Matches', fontsize=8)
plt.xticks(fontsize=8)
plt.title('Total Number of No Result Matches', fontsize=10)
for i,v in enumerate(teams_data['NR']):
    plt.text(i,v,v,fontsize=8)

plt.style.use('ggplot')
plt.show()
```

```python
#Series of Winning percentage
teams_win_per = (teams_data['Win']/teams_data['TotalMatches']) * 100

#Plotting the teams based on the winning percentage
teams_win_per.plot.bar(width=0.5, rot=30)
plt.xlabel('Teams')
plt.ylabel('Winning Percentage')
plt.title('Who has the more winning percentage(Min. 100 Matches)')
plt.style.use('ggplot')
for i,v in enumerate(teams_win_per):
    plt.text(i,v,round(v), fontsize='8')
plt.show()
```

```python
#Team with maximum and minimum winning percentage

for i in range(0,len(teams_win_per)):
    if teams_win_per[i] == teams_win_per.max() :
        index_max = teams_win_per.index[i]
        per_max = round(teams_win_per[i])
print('\nHighest Winning Percentage Team')
print(index_max, per_max)

for i in range(0,len(teams_win_per)):
    if teams_win_per[i] == teams_win_per.min() :
        index_min = teams_win_per.index[i]
        per_min = round(teams_win_per[i])
print('\nLowest Winning Percentage Team')
print(index_min, per_min)

#Finding Top3 Teams with maximum winning percentage and  bottom 3 Teams with winning percentage
print('\nBottom 3 teams with Lowest Winning Percentage')
teams_win_per_sorted = teams_win_per.sort_values(ascending=True)
print(teams_win_per_sorted[:3])
print('\nTop 3 teams with Highest Winning Percentage')
teams_win_per_sorted = teams_win_per.sort_values(ascending=False)
print(teams_win_per_sorted[:3])

#Finding Teams with Most Tied/NR matches
team_tie_nr = teams_data1['Tied/NR'].sort_values()
print("\nTeams with Maximum and Minimum Tie/No Result Matches")
print(team_tie_nr.index[0], team_tie_nr[0])
print(team_tie_nr.index[-1],team_tie_nr[-1])
```

# Analyzing the Highest and Lowest Scores of the Teams

```python
#Creating New DataFrame Containing 4 columns: Highet Score Batting, Lowest Score Batting Highest Score Bowling, Lowest Score Bowling
team_batting_score = teams_data[['HighestScoredRun(Inning)','LowestScoredRun(Inning)', 'HighestConcededRun(Inning)', 'LowestConcededRun(Inning)']]
team_batting_score.columns = ['MaxScore','MinScore','MaxScoreConcd','MinScoreConcd']
```

```python
#plotting the highest and lowest scores scored by the teams
team_batting_score[['MaxScore','MinScore']].plot.bar(rot = 30)
plt.xlabel('Teams')
plt.ylabel('Scores')
plt.title('Who Scored Most vs Who Scored Least in an Inning')
plt.legend(loc='center')
for i,v in enumerate(team_batting_score['MaxScore']):
    plt.text(i,v,str(v))
for i,v in enumerate(team_batting_score['MinScore']):
    plt.text(i,v, str(v))
plt.show()
```

```python
#plotting the highest and lowest scores conceded by the teams
team_batting_score[['MaxScoreConcd','MinScoreConcd']].plot.bar(rot = 30)
plt.xlabel('Teams')
plt.ylabel('Scores')
plt.title('Who Conceded Most vs Who Conceded Least in an Inning')
plt.legend(loc='center')
for i,v in enumerate(team_batting_score['MaxScoreConcd']):
    plt.text(i,v,str(v))
for i,v in enumerate(team_batting_score['MinScoreConcd']):
    plt.text(i,v, str(v))
plt.show()
```

# Analyzing the Average Runs Scored/Conceded by the teams

```python
#Analyzing the average runs scored by the teams
avg_runs_wkt = teams_data['Avg.RunsScored/Wicket']
avg_runs_over = teams_data['Avg.RunsScored/Over']
```

```python
#Plotting top 5 teams with highest avg runs/wkt
avg_runs_wkt = avg_runs_wkt.sort_values(ascending=False)
avg_runs_wkt_5 = avg_runs_wkt[0:5]

avg_runs_wkt_5.plot.bar(rot=30, width=0.4, color='lightblue')
plt.xlabel('Teams')
plt.ylabel('Avg Runs/Wicket')
plt.title('Top 5 teams - Avg. Runs Scored/Wicket')
for i,v in enumerate(avg_runs_wkt_5):
    plt.text(i,v,v)
plt.show()

#plotting top 5 teams with highest avg runs/wkt
avg_runs_over = avg_runs_over.sort_values(ascending=False)
avg_runs_over_5 = avg_runs_over[:5]

avg_runs_over_5.plot.bar(rot=30, width=0.4, color='lightgreen')
plt.xlabel('Teams')
plt.ylabel('Avg Runs/Over')
plt.title('Top 5 teams - Avg. Runs Scored/Over')
for i,v in enumerate(avg_runs_over_5):
    plt.text(i,v,v)

plt.show()
```

```python
avg_runs_conc_wkt = teams_data['Avg.RunsConceded/Wicket']
avg_runs_conc_over = teams_data['Avg.RunsConceded/Over']
```

```python
#Plotting top 5 teams with highest avg runs conceded/wkt
avg_runs_conc_wkt = avg_runs_conc_wkt.sort_values(ascending=False)
avg_runs_conc_wkt_5 = avg_runs_conc_wkt[0:5]

avg_runs_conc_wkt_5.plot.bar(rot=30, width=0.4, color='lightblue')
plt.xlabel('Teams')
plt.ylabel('Avg Runs Conceded/Wicket')
plt.title('Top 5 teams - Avg. Runs Conceded/Wicket')
for i,v in enumerate(avg_runs_conc_wkt_5):
    plt.text(i,v,v)
plt.show()

#plotting top 5 teams with highest avg runs/wkt
avg_runs_conc_over = avg_runs_conc_over.sort_values(ascending=False)
avg_runs_conc_over_5 = avg_runs_conc_over[:5]

avg_runs_conc_over_5.plot.bar(rot=30, width=0.4,color='lightgreen')
plt.xlabel('Teams')
plt.ylabel('Avg Runs Conceded/Over')
plt.title('Top 5 teams - Avg. Runs Conceded/Over')
for i,v in enumerate(avg_runs_conc_over_5):
    plt.text(i,v,v)

plt.style.use('ggplot')
plt.show()
```
